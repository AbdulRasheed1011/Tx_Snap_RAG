```python
from bs4 import BeautifulSoup
```

```python
with open('data/raw/www.hhs.texas.gov_handbooks_texas-works-handbook.html', 'r', encoding = 'utf-8') as f:
    html = f.read()
    soup = BeautifulSoup(html, 'html.parser')
```

```python
for tag in soup(['script', 'style', 'nav', 'footer', 'header']):
    tag.decompose()


lines = []
for element in soup.find_all(["h1", "h2", "h3", "p", "li"]):
        text = element.get_text(strip=True)
        if not text:
            continue

        # Mark headings clearly (important for chunking later)
        if element.name in ["h1", "h2", "h3"]:
            lines.append(f"\n{text.upper()}\n")
        else:
            lines.append(text)

    # 5. Join everything into one clean string
clean_text = "\n".join(lines)

```
